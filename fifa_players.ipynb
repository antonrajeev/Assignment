```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
players=pd.read_csv('fifa_players.csv')
players.head(491)
```

```python
players.isnull().sum()
```

```python
Players=players.dropna()
```

```python
Players.isnull().sum()
```

```python
top_level_players=Players.iloc[0:500,0:40]
```

```python
top_level_players
```

```python
plt.hist(top_level_players['height_cm'])
plt.xlabel("Height(cm)")
plt.ylabel("Frequency")
plt.title("Height distribution of top level players")
```

```python
plt.scatter(Players['overall_rating'],Players['value_euro'],color='red')
plt.xlabel('Overall rating ')
plt.ylabel('Market value')
plt.title('Overall rating vs Market value')
plt.show()
```
